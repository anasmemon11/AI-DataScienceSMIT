```python

import pandas as pd
```

```python

product_data = pd.read_csv("products.csv")
order_data = pd.read_csv("orders.csv")
customer_data = pd.read_csv("customers.csv")

repeating_customers = order_data['customer_id'].value_counts()[order_data['customer_id'].value_counts() > 1].index.tolist()

product_sales = order_data['product_id'].value_counts()
top_sell_pro = product_sales.idxmax()
least_sell_pro = product_sales.idxmin()

more_5_vis = order_data['customer_id'].value_counts()[order_data['customer_id'].value_counts() > 5].index.tolist()

order_data['total_price'] = product_data['price'].astype(float) * order_data['quantity'].astype(int)
tot_inc = order_data['total_price'].sum()


print(f"""
Repeated Customers
-----------------------
{repeating_customers}

Top Saled Products
---------------------
{top_sell_pro}

Least Saled Products
-----------------------
{least_sell_pro}

More than 5 times visitors
-----------------------------
{more_5_vis}

Total income
--------------
{tot_inc}
""")
```
